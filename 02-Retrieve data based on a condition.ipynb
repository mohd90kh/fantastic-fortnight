## 2: HOW TO USE LOC (location)

```python
#one condition

df.loc[ df['Name'] == "Moran, Mr. James" ]
```

```python
#Multiple conditions.

df.loc[ (df['Age'] == 80 ) & (df["Survived"] == 1 )]
```

## 3: HOW TO USE query

```python
#one conditions.

df.query('Sex == "male" and Age < 5 ')
```

```python
#df.loc[df['Age'] == 22 , 'Age'] = 0
#df.loc[df['Age'] == 38 , 'Age'] = 0
df.loc[df['Age'] <= 5 , 'Age'] = 0
df.loc[df['Pclass'] == 3 , 'Pclass'] = 'High'
df.loc[df['Pclass'] == 1 , 'Pclass'] = 'Low'
df.loc[df['Pclass'] == 2 , 'Pclass'] = 'Medium'
```
